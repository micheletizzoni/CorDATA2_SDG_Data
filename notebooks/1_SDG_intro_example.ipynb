# Data selection
We would like to focus on data available for:
- all ages
- all area types
- both genders

```python
df_work[(df_work['[Age]']=='ALLAGE')]
```

Since only one country has data available for all ages we will compute averages across age groups

```python
df_selection=df_work[(df_work['[Location]']=='ALLAREA')]
```

```python
df_selection['GeoAreaName'].unique().shape[0]
```

```python
df_selection.head()
```

### When was the most recent survey conducted in each country?

```python
df=df_selection[['GeoAreaName','TimePeriod']].groupby(['GeoAreaName']).max().reset_index()
```

```python
df['TimePeriod'].hist(bins=18, xlabelsize=18, ylabelsize=18, figsize=(10,7))
plt.xticks(np.arange(2000, 2019, step=2))
plt.xlabel('Year', fontsize=18)
plt.ylabel('N. countries', fontsize=18)
plt.show()
```

### We select the entries for the most recent survey in each country

```python
df_sel = df_selection[df_selection.set_index(['GeoAreaName','TimePeriod']).index.isin(df.set_index(['GeoAreaName','TimePeriod']).index)]
```

```python
df_sel.head()
```

# Computing a gender ratio value for this indicator

```python
df_gender=df_sel[['GeoAreaName','[Sex]','TimePeriod','Value']].groupby(['GeoAreaName','[Sex]']).mean()
```

```python
df_gender
```

```python
df_f=df_gender.xs('FEMALE', level='[Sex]')
df_f=df_f.rename(columns={'Value':'Value_F'})
df_f.reset_index(inplace=True)
```

```python
df_m=df_gender.xs('MALE', level='[Sex]')
df_m=df_m.rename(columns={'Value':'Value_M'})
df_m.reset_index(inplace=True)
```

```python
df=df_m.merge(df_f, how='outer')
```

```python
df
```

```python
df['Gender_ratio']=df['Value_F']/df['Value_M']
```

```python
df['Gender_ratio'].hist(bins=81,xlabelsize=18, ylabelsize=18, figsize=(10,7))
plt.xlabel('Gender ratio', fontsize=18)
plt.ylabel('N. countries', fontsize=18)
plt.show()
```

```python
df_sorted=df.sort_values('Gender_ratio', ascending=False)
```

# Final chart
Finally, we make a chart of our result that can be used for publication

```python
f, ax = plt.subplots(figsize=(8, 15))
sns.set(style="whitegrid")

sns.barplot(x='Gender_ratio', y='GeoAreaName', data=df_sorted, label="Ratio", palette='inferno')
plt.xticks(fontsize=14)
plt.yticks(fontsize=12)
plt.xlabel('Female to male ratio of time spent in upaid care work', fontsize=18)
plt.ylabel('')
```
